```python
import pandas as pd
df = pd.read_csv("hotel_bookings.csv")
df.head()
df.info()
```

```python
df = df.drop(columns=['company'], errors='ignore')
```

```python
df['children'] = df['children'].fillna(0)
df['country'] = df['country'].fillna('Unknown')
df['agent'] = df['agent'].fillna(0)

```

```python
df.isnull().sum()
```

```python
df = df[[
    'hotel',
    'arrival_date_month',
    'stays_in_weekend_nights',
    'stays_in_week_nights',
    'adults',
    'children',
    'adr',
    'market_segment',
    'customer_type',
    'is_canceled'
]]
```

```python
def budget_category(price):
    if price < 80:
        return 'Low'
    elif price < 150:
        return 'Medium'
    else:
        return 'High'

df['budget_category'] = df['adr'].apply(budget_category)
```

```python
def budget_category(price):
    if price < 80:
        return 'Low'
    elif price < 150:
        return 'Medium'
    else:
        return 'High'

df['budget_category'] = df['adr'].apply(budget_category)

df['budget_category'].value_counts()
```

```python
import numpy as np

destinations = ['Beach', 'City', 'Hill', 'Wildlife']
df['destination_type'] = np.random.choice(destinations, size=len(df))

df['destination_type'].value_counts()
```

```python
df['demand_level'] = df['is_canceled'].map({0: 'High', 1: 'Low'})

df['demand_level'].value_counts()
```

```python
df.head()
```

```python
df_final = df[[
    'arrival_date_month',
    'stays_in_weekend_nights',
    'stays_in_week_nights',
    'travel_type',
    'children',
    'budget_category',
    'destination_type',
    'demand_level'
]]
```

```python
df_final.info()
df_final.head()
```

```python
df_final = df_final.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for col in df_final.columns:
    df_final[col] = encoder.fit_transform(df_final[col])
```

```python
df_final.to_csv("final_dataset.csv", index=False)
```
